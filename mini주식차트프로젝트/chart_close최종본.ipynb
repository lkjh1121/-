```python
facebook = pd.read_csv("./fb2.csv")
facebook
```

```python
google = pd.read_csv("./googl2.csv")
google
```

```python
netflix = pd.read_csv("./nflx2.csv")
netflix
```

```python
apple_sub = apple[['date', 'mean_close']]
apple_sub.rename( columns={'mean_close':'Apple_mean_close'}, inplace=True)

amazon_sub = amazon[['date', 'mean_close']]
amazon_sub.rename( columns={'mean_close':'Amazon_mean_close'}, inplace=True)

facebook_sub = facebook[['date', 'mean_close']]
facebook_sub.rename( columns={'mean_close':'Facebook_mean_close'}, inplace=True)

google_sub = google[['date', 'mean_close']]
google_sub.rename( columns={'mean_close':'Google_mean_close'}, inplace=True)

netflix_sub = netflix[['date', 'mean_close']]
netflix_sub.rename( columns={'mean_close':'Netflix_mean_close'}, inplace=True)
all = pd.merge(apple_sub, amazon_sub, how="outer", on='date')
all = pd.merge(all, facebook_sub, how="outer", on='date')
all = pd.merge(all, google_sub, how="outer", on='date')
all = pd.merge(all, netflix_sub, how="outer", on='date')
all
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import set_matplotlib_formats 

set_matplotlib_formats('retina')    # 글씨 선명하게


# 한글 폰트 설정 전에 스타일을 지정해야한다.
sns.set(color_codes=True)
sns.set_style('darkgrid')

###### 차트에 환경설정 : 한글 폰트 지정하기
plt.rcParams['font.family']='Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

# sns.lineplot(data = all,)
# all.plot()
x = all.loc[:,"date"]

# all.columns
plt.plot(x, all.loc[:,"Apple_mean_close"], label='apple', color='r', marker='o') 
plt.plot(x, all.loc[:,"Amazon_mean_close"], label='Amazon', color='y', marker='o') 
plt.plot(x, all.loc[:,"Facebook_mean_close"], label='Facebook', color='black', marker='o') 
plt.plot(x, all.loc[:,"Google_mean_close"], label='Google', color='g', marker='o') 
plt.plot(x, all.loc[:,"Netflix_mean_close"], label='Netflix', color='b', marker='o')  

plt.rcParams['figure.figsize'] = [15, 8] # [width, height] (inches)
plt.xticks(rotation=45)

plt.legend(loc="best") 

plt.show()
```
